```python
import xarray as xr
```

```python
ds = xr.open_dataset('../data/raw/moc_transports_mod.nc')
ds = xr.concat([ds for _ in range(100)],dim='time')
ds
```

```python
dsDask = ds.chunk({'time':35000})
%timeit -n 1 -r 1 dsDask.mean(dim='time')
```

```python
df = ds.to_dataframe()
df.shape
```

```python
%timeit -n 1 -r 1 dsDask.mean(dim='time')
```
